```python
import numpy as np 
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/swapnasavalgi/Marketing-Data/main/marketing_data.csv'
data = pd.read_csv(url)
data
```

```python
missing_values = data.isnull().sum()
missing_values
```

```python
data.info()
```

There are 5 categorical variables and 23 numerical variables

```python
#Removing the space from "Income" var

data.rename(columns={' Income ':'Income'},inplace=True)
data.info()
```

```python
data["Income"] = data["Income"].str.replace("$", "").str.replace(",","")
data['Income'] = data['Income'].astype('float')
```

```python
#Adding a new variable to replace year_birth to Age. 
data['Age'] = 2021 - data['Year_Birth']
```

```python
#Combining below variables into one.

data['Spending'] = data['MntWines'] + data['MntFruits']+data['MntMeatProducts']+data['MntFishProducts']+data['MntSweetProducts']+data['MntGoldProds']
```

```python
data['Marital_Status'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data['Marital_Status'] = data['Marital_Status'].replace({'Together':'In Couple','Married':'In Couple', 'Single':'Alone', 'Widow':'Alone','Divorced':'Alone','Absurd':'Alone','YOLO':'Alone' })
```

```python
data['Marital_Status'].value_counts()
```

```python
data['Has_child'] = np.where(data['Kidhome'] + data['Teenhome'] > 0 , 'Yes','No')
```

```python
data['Education'].value_counts()
```

```python
#Variable Educationnal_years as the total number of years of education the individual achieved according to its diploma

data['Educational_years'] = data['Education'].replace({'Basic':5,'2n Cycle':8,'Graduation':12,'Master':18,'PhD':21})
```

```python
#Creting an new dataaframe

data_new = data[['Age','Income','Spending','Marital_Status','Has_child','Educational_years', 'Education']]
data_new.head()
```
